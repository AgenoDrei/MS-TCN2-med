```python
def check_start_stop_time(start, stop):
    if pd.isnull(start) or pd.isnull(stop):
        return True
    if len(start)>8 or len(stop)>8:
        return False
    
    try:
        start_min = start.split(':')[0]
        start_sec = start.split(':')[1]
    
        stop_min = stop.split(':')[0]
        stop_sec = stop.split(':')[1]
    except:
        return False
    
    if int(start_min) < int(stop_min):
        return True
    elif int(start_min) == int(stop_min) and int(start_sec) < int(stop_sec):
        return True
    return False

# Manually inspect the annotations, row by row
step_size = 1
for i, row in annos.iterrows():
    row_string = ' '.join([str(value) for value in row[2::step_size]])
    print(row['Video ID'] + ': ' + row_string)
    #print(len(row))
    
    for i in range(2, len(row) - 3, 2):
        start = row.iloc[i]
        stop = row.iloc[i+1]
        
        if not check_start_stop_time(start, stop):
            print(f'Error in {row["Video ID"]}, start: {start}, stop: {stop}')
            break
    
```

```python
#annos[0:40]
#annos[40:80]

def check_format(time):
    if pd.isnull(time):
        return time
    try:
        value_min = time.split(':')[0]
        value_sec = time.split(':')[1]
    except:
        return time
    
    return value_min + ':' + value_sec

annos.iloc[:, 2:] = annos.iloc[:, 2:].applymap(check_format)
```
